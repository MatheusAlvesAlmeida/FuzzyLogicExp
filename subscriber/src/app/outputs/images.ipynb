```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
hpa = pd.read_csv('hpa_4500.csv')
ga1 = pd.read_csv('ga_round1_4500.csv')
ga2 = pd.read_csv('ga_round2_4500.csv')
ga3 = pd.read_csv('ga_round3_4500.csv')
default = pd.read_csv('default_rules_4500.csv')
```

```python
def plot_setpoint_vs_current_arrival_rate(df, title):
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['setpoint'], label='Setpoint', linestyle='--')
    plt.plot(df.index, df['current_arrival_rate'], label='Current Arrival Rate', marker='o')
    plt.xlabel('Index')
    plt.ylabel('Rate')
    plt.title(f'{title}')
    plt.legend()
    plt.grid(True)
    
    # RMSE calculation
    rmse = np.sqrt(mean_squared_error(df['setpoint'], df['current_arrival_rate']))
    
    # Display RMSE on plot
    plt.text(0.95, 0.1, f'RMSE: {rmse:.2f}', horizontalalignment='right', verticalalignment='center', 
             transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
    
    # Show plot
    plt.show()
    
    return rmse
```

```python
plot_setpoint_vs_current_arrival_rate(hpa, 'HPA Controller')
```

```python
plot_setpoint_vs_current_arrival_rate(default, 'Fuzzy Controller - Default Rules')
```

```python
plot_setpoint_vs_current_arrival_rate(ga1, 'Fuzzy Controller - Rules improved by GA Round 1')
```

```python
plot_setpoint_vs_current_arrival_rate(ga2, 'Fuzzy Controller - Rules improved by GA Round 2')
```

```python
plot_setpoint_vs_current_arrival_rate(ga3, 'Fuzzy Controller - Rules improved by GA Round 3')
```
